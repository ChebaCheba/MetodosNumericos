```python
import numpy as np
import matplotlib.pyplot as plt

def q(y):
    return 1-((20**2)*(3+y))/(9.81*(3*y+(y**2)/2)**3)

#Elegir valores iniciales x0 y x1
#Donde haya un cambio de signo

yarray = np.linspace(10,25,100)
qarray = np.zeros(100)

for i in range(100):
    qarray[i] = q(yarray[i])
plt.plot(qarray,yarray)
plt.grid()
```
